 **4 Explain how props support React one way data flow?**
 --In React Props are used to pass data one component to another component. Props help implemet react one-way -data flow system.

 One -way-dat flow system means that data in React application flows in a single direction from parent component to child component.

 How props  support one way data flow:
 1. Data flows from parent to child :
  in react a parant componenet passed data to child component using props.The child component recieves this data and displays it but it cannot change the original value.

  example:
  parent component:
  <greeting name="shalini"/>

  child component:
  function Greeting(props){
    return <h1> hellow,{props.name}!</h1>;}


  2  props are Read only-
    props cannot be modified inside the child component.
    
3 Better Data control-Since only the parent component can update the data, it become easier to manage.

4 Easier dibbugging:When data flows in one direction it is easier to identify the where changes are happening, making dibbugging simpler.

**7 Why is the key atttriibute  important rendering lists in React? what goes wrong if keys are missing or imssuse array index?**
-- In React the key attribute is used to uniquely indentify elements in a list. when rendering multiple elements  using the map() meathod react reuire a key for each element.

The key  helps rect efficiently update and manage changes in a user interface. React uses a proces called reconcilation  to comapre the Virtual DOM with the ral DOM. During this process keys helps react identify which item have been added removed or update.

**Importance of key attribute**
. Efficient update:Keys allows React to update only the changed element insted of re-rendering the entire list.

.Better Performrnce: Proper keys improve rendering performrence.

.Stable indentity: Each element gets a unique identity ,which help React track changes correctly.

.Prevent UI bugs:  Correct keys prevent unexpected behaviour during re-rendering.

**What happen when keys are missing:**
- React shows a warning message in the console.
